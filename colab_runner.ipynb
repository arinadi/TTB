2.  **Runtime**: Change runtime type to **GPU** (Runtime > Change runtime type > T4 GPU).

```python
import os
import sys
from google.colab import userdata

# --- CONFIGURATION ---
# ‚ö†Ô∏è REPLACE THIS WITH YOUR REPOSITORY URL
REPO_URL = "https://github.com/arinadi/TTB.git" # Example
REPO_NAME = "TTB"
# ---------------------

try:
    # Try to use GITHUB_TOKEN if available for private repos
    token = userdata.get('GITHUB_TOKEN')
    if token and "github.com" in REPO_URL:
        REPO_URL = REPO_URL.replace("https://", f"https://{token}@")
except Exception:
    pass

if not os.path.exists(REPO_NAME):
    print(f"‚è≥ Cloning {REPO_NAME}...")
    !git clone {REPO_URL}
    %cd {REPO_NAME}
else:
    print(f"‚è≥ Updating {REPO_NAME}...")
    %cd {REPO_NAME}
    !git pull

print("‚úÖ Code updated.")
```
